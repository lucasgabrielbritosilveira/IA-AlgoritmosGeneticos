# Análise do experimento 0

```python
import pandas as pd

caminho_csv = './resultados_gaoriginal.csv'

df = pd.read_csv(caminho_csv)

df
```

### Verificando as combinações que chegaram ao melhor custo (menos resultado)

```python
# Encontrar o menor custo
menor_custo = df['melhor_custo'].min()

# Filtrar os registros com o menor custo
df_melhor_resultado = df[df['melhor_custo'] == menor_custo]

print(f"Menor valor de 'melhor_custo': {menor_custo}")
df_melhor_resultado
```

### Verificando quais combinações chegaram ao menor custo mais rápido

```python
menor_geracao = df_melhor_resultado['geracao_melhor_custo'].min()

# Filtrar os registros que chegaram ao melhor resultado mais rápido
df_mais_rapido = df_melhor_resultado[df_melhor_resultado['geracao_melhor_custo'] == menor_geracao]

# Ordenar pelos que demoraram menos em tempo de execução para um critério extra
df_mais_rapido = df_mais_rapido.sort_values(by='tempo_execucao_s')

# Exibir os resultados filtrados
print(f"Menor geração onde o melhor custo foi alcançado: {menor_geracao}")
df_mais_rapido
```

```python
# Agrupar por tamanho da população
pop_analysis = df.groupby('pop_size').agg({
    'melhor_custo': 'mean',
    'tempo_execucao_s': 'mean',
    'geracao_melhor_custo': 'mean'
}).reset_index()

pop_analysis.rename(columns={
    'melhor_custo': 'Custo Médio',
    'tempo_execucao_s': 'Tempo Médio (s)',
    'geracao_melhor_custo': 'Geração Média p/ Melhor Custo'
}, inplace=True)

pop_analysis
```

```python
import matplotlib.pyplot as plt

# Plot do custo médio
plt.figure()
plt.plot(pop_analysis['pop_size'], pop_analysis['Custo Médio'], marker='o')
plt.title('Custo Médio vs Tamanho da População')
plt.xlabel('Tamanho da População')
plt.ylabel('Custo Médio')
plt.grid(True)
plt.show()

# Plot do tempo médio
plt.figure()
plt.plot(pop_analysis['pop_size'], pop_analysis['Tempo Médio (s)'], marker='o')
plt.title('Tempo Médio vs Tamanho da População')
plt.xlabel('Tamanho da População')
plt.ylabel('Tempo Médio (s)')
plt.grid(True)
plt.show()

# Plot da geração média em que o melhor custo foi alcançado
plt.figure()
plt.plot(pop_analysis['pop_size'], pop_analysis['Geração Média p/ Melhor Custo'], marker='o')
plt.title('Geração Média p/ Melhor Custo vs Tamanho da População')
plt.xlabel('Tamanho da População')
plt.ylabel('Geração Média p/ Melhor Custo')
plt.grid(True)
plt.show()
```
